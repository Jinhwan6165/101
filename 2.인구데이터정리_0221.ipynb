```python
# 인구데이터 "서울특별시 종로구 (1111000000)"  --> "서울특별시 종로구"
# 치킨데이터 "서울특별시"   "종로구"     --> "서울특별시 종로구"
```

```python
temp_list = []
for c in raw['행정기관']:
    #print(c[:-13])   # 글자수로 정리 하는 방법
    temp=c.split(' (')[0]    # ( 를 앞뒤로 나누어서 정리하는 방법
    print (temp)
    # 코드 정리 방법
    ###  길이로
    ### "(" 기준으로 데이터 split --> 선택
    temp_list.append(temp)
temp_list
```

```python
raw['시군구정리'] = temp_list
raw.head()
```

```python
raw['시군구정리'].unique()
```

```python
raw['시군구정리'].value_counts()
# 세종특별자치시에는 왜 동일한게 2개 들어가 있는가?
# 18년도 말 기준으로는 큰/작은 카테고리가 동일하기 때문에 2개로 count 됨
```
