```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
browser = webdriver.Chrome('C:/informs/Chromedriver.exe')
```

```python
results = []
```

```python
url = 'https://www.melon.com/chart/index.htm'
browser.get(url)

html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
song_list = soup.select('div > table > tbody > tr')
song = song_list[0]

for song in song_list:
    title = song.select('div > div.ellipsis.rank01 > span > a')[0].text.strip()
    singer = song.select('div > div.ellipsis.rank02 > span > a')[0].text.strip()
    album = song.select('div > div.rank03 > a')[0].text.strip()
    like = song.select('td > div > button > span.cnt')[0].text.strip()
    data = [title, singer, album, like]
    len(like)
    print(like)
    #results.append(data)
```
