```python
# 업종중분류 ' 닭/오리요리'인 데이터 선택
cond =  raw['업종중분류'] == '닭/오리요리'
df = raw[cond]
df
```

```python
df['계'] = df.sum(axis = 1)  #컬럼별로 더하기
df.head()
```

```python
# 치킨집 수 합계 구하기.
#df['컬럼명1'] + df['컬럼명2']+...
df.describe
```
